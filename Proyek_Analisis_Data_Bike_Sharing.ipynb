## Visualization & Explanatory Analysis

Berapa persen perbedaan rata-rata jumlah pengguna sepeda pada hari kerja dan akhir pekan?

```python
# Menghitung rata-rata jumlah pengguna sepeda pada hari kerja
mean_weekday = df_day[df_day['workingday'] == 1]['cnt'].mean()

# Menghitung rata-rata jumlah pengguna sepeda pada akhir pekan
mean_weekend = df_day[df_day['workingday'] == 0]['cnt'].mean()

# Menghitung perbedaan antara kedua rata-rata tersebut
difference = mean_weekday - mean_weekend

# Menghitung persentase perbedaan
percentage_difference = (difference / mean_weekday) * 100

# Membuat visualisasi
labels = ['Weekday', 'Weekend']
values = [mean_weekday, mean_weekend]

plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['blue', 'orange'])
plt.title('Perbandingan Rata-Rata Jumlah Pengguna Sepeda\nAntara Hari Kerja dan Akhir Pekan')
plt.xlabel('Hari')
plt.ylabel('Rata-Rata Jumlah Pengguna Sepeda')
plt.text(0, mean_weekday, f'{mean_weekday:.2f}', ha='center', va='bottom')
plt.text(1, mean_weekend, f'{mean_weekend:.2f}', ha='center', va='bottom')
plt.text(0.5, max(mean_weekday, mean_weekend), f'Perbedaan: {difference:.2f}\nPersentase: {percentage_difference:.2f}%', ha='center', va='bottom')
plt.ylim(0, max(mean_weekday, mean_weekend) * 1.2)
plt.show()
```

**Insight**

ini menunjukkan bahwa rata-rata jumlah pengguna sepeda pada akhir pekan lebih tinggi daripada pada hari kerja. Hal ini mungkin disebabkan oleh waktu luang yang lebih banyak pada akhir pekan, sehingga orang lebih cenderung menggunakan sepeda untuk rekreasi atau kegiatan outdoor.

Apakah terdapat korelasi positif antara temperatur dan jumlah pengguna sepeda pada musim panas (Juni - Agustus) di tahun 2011?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca data
df_day = pd.read_csv("day.csv")

# Filter data untuk musim panas (Juni - Agustus) di tahun 2011
summer_2011 = df_day[(df_day['yr'] == 0) & (df_day['mnth'] >= 6) & (df_day['mnth'] <= 8)]

# Hitung korelasi antara temperatur dan jumlah pengguna sepeda
correlation_temp_cnt = summer_2011[['temp', 'cnt']].corr().iloc[0, 1]

# Membuat visualisasi
plt.figure(figsize=(8, 6))
sns.scatterplot(data=summer_2011, x='temp', y='cnt')
plt.title('Korelasi Antara Temperatur dan Jumlah Pengguna Sepeda\nMusim Panas 2011')
plt.xlabel('Temperatur (C)')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.text(min(summer_2011['temp']), max(summer_2011['cnt']), f'Korelasi: {correlation_temp_cnt:.2f}', ha='left', va='top', fontsize=12)
plt.grid(True)
plt.show()
```

Bagaimana perkembangan jumlah penyewaan sepeda dari bulan ke bulan?

```python
# Mengelompokkan data berdasarkan bulan dan menghitung total penyewaan sepeda setiap bulannya
monthly_rentals = df_day.groupby('mnth')['cnt'].sum()

# Membuat visualisasi
plt.figure(figsize=(10, 6))
monthly_rentals.plot(marker='o', color='b', linewidth=2, markersize=8)
plt.title('Perkembangan Jumlah Penyewaan Sepeda dari Bulan ke Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

Bagaimana korelasi antara suhu dan jumlah penyewaan sepeda pada hari-hari kerja dan hari-hari libur?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca data
df_day = pd.read_csv("day.csv")

# Memfilter data untuk hari-hari kerja dan hari-hari libur
workdays = df_day[df_day['workingday'] == 1]
holidays = df_day[df_day['workingday'] == 0]

# Membuat visualisasi korelasi antara suhu dan jumlah penyewaan sepeda pada hari-hari kerja
plt.figure(figsize=(8, 6))
sns.scatterplot(data=workdays, x='temp', y='cnt')
plt.title('Korelasi antara Suhu dan Jumlah Penyewaan Sepeda pada Hari Kerja')
plt.xlabel('Suhu (C)')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.grid(True)
plt.show()

# Membuat visualisasi korelasi antara suhu dan jumlah penyewaan sepeda pada hari-hari libur
plt.figure(figsize=(8, 6))
sns.scatterplot(data=holidays, x='temp', y='cnt')
plt.title('Korelasi antara Suhu dan Jumlah Penyewaan Sepeda pada Hari Libur')
plt.xlabel('Suhu (C)')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.grid(True)
plt.show()
```

**Insight**

`Perbedaan Antara Hari Kerja dan Hari Libur:`

Jika terdapat perbedaan signifikan dalam korelasi antara suhu dan jumlah penyewaan sepeda antara hari kerja dan hari libur, ini menunjukkan bahwa faktor-faktor lain mungkin memengaruhi hubungan antara suhu dan penyewaan sepeda tergantung pada apakah itu hari kerja atau hari libur.

Bagaimana tren keseluruhan jumlah penyewaan sepeda selama periode yang diamati?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data
df_day = pd.read_csv("day.csv")

# Mengubah kolom 'dteday' menjadi tipe data datetime
df_day['dteday'] = pd.to_datetime(df_day['dteday'])

# Menghitung total jumlah penyewaan sepeda per hari
daily_rentals = df_day.groupby('dteday')['cnt'].sum()

# Membuat visualisasi
plt.figure(figsize=(12, 6))
daily_rentals.plot(color='b', linewidth=2)
plt.title('Tren Keseluruhan Jumlah Penyewaan Sepeda')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.axhline(daily_rentals.mean(), color='r', linestyle='--', label=f'Rata-rata: {daily_rentals.mean():.2f}')
plt.legend()
plt.grid(True)
plt.show()
```
