## Visualization & Explanatory Analysis

Berapa persen perbedaan jumlah orang yang menggunakan sepeda di hari kerja versus akhir pekan?***

Metode untuk menghitung perbedaan antara jumlah orang yang menggunakan sepeda di hari kerja dan akhir pekan:

1. Kelompokkan data berdasarkan hari kerja, dengan nol untuk akhir pekan dan satu untuk hari kerja.
2. Tentukan jumlah pengguna sepeda rata-rata untuk setiap kelompok.
3. Tentukan perbedaan persentase antara rata-rata jumlah pengguna sepeda pada hari kerja dan akhir pekan.

```python
avg_cnt_by_workingday = df_day.groupby('workingday')['cnt'].mean()
diff_percentage = ((avg_cnt_by_workingday[1] - avg_cnt_by_workingday[0]) / avg_cnt_by_workingday[0]) * 100

print("Perbedaan rata-rata jumlah pengguna sepeda pada hari kerja dan akhir pekan: {:.2f}%".format(diff_percentage))
```

**Apakah terdapat korelasi positif antara temperatur dan jumlah pengguna sepeda pada musim panas (Juni - Agustus) di tahun 2011?**
1. Filter data untuk musim panas (Juni - Agustus) di tahun 2011.
2. Hitung korelasi antara variabel temperatur dan jumlah pengguna sepeda pada data yang difilter.

```python
summer_2011 = df_day[(df_day['yr'] == 0) & (df_day['mnth'].isin([6, 7, 8]))]
correlation_temp_cnt = summer_2011[['temp', 'cnt']].corr().iloc[0, 1]

print("Korelasi antara temperatur dan jumlah pengguna sepeda pada musim panas di tahun 2011:", correlation_temp_cnt)
```
